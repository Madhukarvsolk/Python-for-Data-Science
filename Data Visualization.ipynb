```python
#Line PLot
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=np.arange(0,10,0.1)
y=2*x+5

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,10,1)
y=2*x+5

fig=plt.figure(figsize=(10,5))

plt.plot(x,y,linewidth=2.0,linestyle=":",color='r',alpha=0.5,marker='^')

plt.title("line plot")
plt.xlabel("X-lable")
plt.ylabel("Y-lable")
plt.legend(['line1'],loc='lower left')
plt.grid(True)
plt.show()
```

```python
#Sub Plots
```

```python
x=np.arange(0,10,1)

y1=2*x+5
y2=3*x+10

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('Graph1')

plt.plot(x,y1,linewidth=2.0,linestyle=":",color='r',alpha=0.5,marker='^')
plt.show()

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph2')

plt.plot(x,y2,linewidth=2.0,linestyle=":",color='r',alpha=0.5,marker='^')
plt.show()
```

```python
#Bar Plot
```

```python
data={'Apple':10, 'Mango':20, 'Banana':30, 'Orange':40}

names=list(data.keys())
values=list(data.values())

fig=plt.figure(figsize=(5,5))
plt.bar(names,values)
plt.show()
```

```python
data={'Apple':10, 'Mango':20, 'Banana':30, 'Orange':40, 'Pineapple':10}

names=list(data.keys())
values=list(data.values())
 
fig=plt.figure(figsize=(10,5))

#Horizontal Bar Plot
#Different Colours 

colors=['red', 'green','yellow', 'orange','yellow']
plt.barh(names,values,color=colors)
plt.title('Bar Chart')
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show()
```

```python
#Scatter Graph
```

```python
import matplotlib.pyplot as plt

a=[1,2,3,4,5,6,7,8]
b=[1,12,13,20,25,22,27,23]
c=[30,34,36,40,44,47,46,12]

plt.scatter(a,b)
plt.scatter(a,c)
plt.legend(['a vs b', 'a vs c'], loc='best')
plt.show()
```

```python
import pandas as pd

a=[1,2,3,4,5,6,7,8]
b=[1,12,13,20,25,22,27,23]
c=[30,34,36,40,44,47,46,12]
e=[5,8,12,14,17,20,23,25]

plt.scatter(a,b,c='g',s=200,facecolor='r',marker='o',alpha=0.5)
plt.scatter(a,c,c='b',s=400,facecolor='b',marker='4',alpha=1)
plt.scatter(a,e,c='y',s=300,facecolor='y',marker='3',alpha=1)


plt.legend(['b','c'],loc='best')
plt.legend(['b','e'],loc='best')

plt.grid(True)

plt.show()
```

```python
#Pie Chart
```

```python
label=['dog','cat','snake','lion']
sizes=[10,50,23,33]

plt.pie(sizes,labels=label, autopct='%1.1f%%', shadow=True, startangle=90, explode=(0,0,0.2,0))

plt.title('Pie Plot')
plt.show()


```

```python
#Box Plot
```

```python
total=[12,34,56,34,44,75,3,14]
order=[2,34,54,13,11,23,4,34]
discount=[10,20,30,40,50,60,70,80]

data=list([total,order,discount])

plt.boxplot(data, showmeans=True)

plt.title('Box Plot')
```
