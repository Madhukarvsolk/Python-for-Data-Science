```python
#DataFrame Merging 

Purusha=pd.DataFrame({'ID':[1,2,3,4,5], 'HUSBAND_NAME':['Krishna','Rama','Shiva','Kalki','Brahma']})

Purusha
```

```python
pd.merge(Purusha,Prakruti,left_on='ID',right_on='ID')
```

```python
pd.merge(Purusha,Prakruti,how='left',on='ID')
```

```python
#Data Cleaning
```

```python
#Column Rename
```

```python
cars=cars.rename(columns={'Unnamed: 1':'mode1'})
cars
```

```python
#Fill Mean in the place of Null values

cars.qsec=cars['qsec'].fillna(cars.qsec.mean())
cars
```

```python
#To Drop Unwanted Columns 

cars=cars.drop(columns=['S.No'])
cars
```

```python
#To find Correlation

df=cars[['qsec','cyl','disp']].corr()
df
```

```python
#To find Corr if data is String, we need to convert it to float or int using this 

cars.mpg=cars.mpg.astype(float)
cars.info(show_counts=True)
```
